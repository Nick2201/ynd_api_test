```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# Define the number of rows to generate
num_rows = 1000

# Generate unique CampaignIds and AdGroupIds
num_campaigns = 23
num_adgroups = 5

campaign_ids = list(range(1, num_campaigns + 1))
adgroup_ids = list(range(1, num_adgroups + 1))

campaign_ids *= num_rows // len(campaign_ids) + 1
adgroup_ids *= num_rows // len(adgroup_ids) + 1

campaign_ids = campaign_ids[:num_rows]
adgroup_ids = adgroup_ids[:num_rows]

# Create Campaigns table
campaigns_data = {
    'CampaignId': campaign_ids,
    'Name': [fake.word() for _ in range(num_rows)],
    'Status': [random.choice(['Active', 'Paused']) for _ in range(num_rows)],
    'Budget': [random.uniform(500, 2000) for _ in range(num_rows)]
}
campaigns_df = pd.DataFrame(campaigns_data)

# Create AdGroups table
ad_groups_data = {
    'AdGroupId': adgroup_ids,
    'CampaignId': random.choices(campaigns_df['CampaignId'], k=num_rows),
    'Name': [fake.word() for _ in range(num_rows)],
    'Status': [random.choice(['Active', 'Paused']) for _ in range(num_rows)],
    'Bid': [random.uniform(0.25, 0.75) for _ in range(num_rows)]
}
ad_groups_df = pd.DataFrame(ad_groups_data)

# Create Ads table
ads_data = {
    'AdId': list(range(1, num_rows + 1)),
    'AdGroupId': random.choices(ad_groups_df['AdGroupId'], k=num_rows),
    'Title': [fake.sentence(nb_words=3) for _ in range(num_rows)],
    'Description': [fake.sentence(nb_words=6) for _ in range(num_rows)],
    'Status': [random.choice(['Active', 'Paused']) for _ in range(num_rows)]
}
ads_df = pd.DataFrame(ads_data)

# Create Keywords table
keywords_data = {
    'KeywordId': list(range(1, num_rows + 1)),
    'AdGroupId': random.choices(ad_groups_df['AdGroupId'], k=num_rows),
    'KeywordText': [fake.word() for _ in range(num_rows)],
    'Status': [random.choice(['Active', 'Paused']) for _ in range(num_rows)],
    'Bid': [random.uniform(0.25, 0.75) for _ in range(num_rows)]
}
keywords_df = pd.DataFrame(keywords_data)

# Create Spend table
spend_data = {
    'Date': pd.date_range(start='2023-06-01', periods=num_rows),
    'CampaignId': random.choices(campaigns_df['CampaignId'], k=num_rows),
    'AdGroupId': random.choices(ad_groups_df['AdGroupId'], k=num_rows),
    'Spend': [random.uniform(100, 1000) for _ in range(num_rows)]
}
spend_df = pd.DataFrame(spend_data)

# Create Statistics table
statistics_data = {
    'Date': pd.date_range(start='2023-06-01', periods=num_rows),
    'CampaignId': random.choices(campaigns_df['CampaignId'], k=num_rows),
    'AdGroupId': random.choices(ad_groups_df['AdGroupId'], k=num_rows),
    'AdId': random.choices(ads_df['AdId'], k=num_rows),
    'Clicks': [random.randint(50, 200) for _ in range(num_rows)],
    'Impressions': [random.randint(500, 2000) for _ in range(num_rows)]
}
statistics_df = pd.DataFrame(statistics_data)

# Create Web Traffic Sources table
web_traffic_data = {
    'Date': pd.date_range(start='2023-06-01', periods=num_rows),
    'CampaignId': random.choices(campaigns_df['CampaignId'], k=num_rows),
    'Source': random.choices(['Organic Search', 'Social Media', 'Referral', 'Paid Search', 'Direct'], k=num_rows),
    'Clicks': [random.randint(50, 200) for _ in range(num_rows)],
    'Impressions': [random.randint(500, 2000) for _ in range(num_rows)]
}
web_traffic_df = pd.DataFrame(web_traffic_data)

# Create Leads table
leads_data = {
    'Date': pd.date_range(start='2023-06-01', periods=num_rows),
    'CampaignId': random.choices(campaigns_df['CampaignId'], k=num_rows),
    'AdGroupId': random.choices(ad_groups_df['AdGroupId'], k=num_rows),
    'AdId': random.choices(ads_df['AdId'], k=num_rows),
    'Leads': [random.randint(1, 10) for _ in range(num_rows)]
}
leads_df = pd.DataFrame(leads_data)

# Create Page Views table
page_views_data = {
    'Date': pd.date_range(start='2023-06-01', periods=num_rows),
    'Page': [fake.word() for _ in range(num_rows)],
    'Views': [random.randint(100, 1000) for _ in range(num_rows)]
}
page_views_df = pd.DataFrame(page_views_data)

# Create Goal Completion Rate table
goal_completion_data = {
    'Date': pd.date_range(start='2023-06-01', periods=num_rows),
    'CampaignId': random.choices(campaigns_df['CampaignId'], k=num_rows),
    'AdGroupId': random.choices(ad_groups_df['AdGroupId'], k=num_rows),
    'AdId': random.choices(ads_df['AdId'], k=num_rows),
    'Goals Completed': [random.randint(1, 5) for _ in range(num_rows)]
}
goal_completion_df = pd.DataFrame(goal_completion_data)
```
